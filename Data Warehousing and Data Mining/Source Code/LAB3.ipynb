```python
# Implementation of Apriori Algorithm and Association Rule Generation

# Prepared By : Arjun Mijar
# Roll No.    : 79010160

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from IPython.display import display_html

# Function to Display Transactions
def view_transactions(dataset):

    print("\n========== TRANSACTION DATASET ==========")

    for i, transaction in enumerate(dataset):
        print(f"T{i+1} : {transaction}")


# Function to Apply Apriori Algorithm
def apply_apriori(dataset):

    print("\n========== APPLYING APRIORI ALGORITHM ==========")

    # Convert Dataset into Transaction Format
    te = TransactionEncoder()
    te_data = te.fit(dataset).transform(dataset)

    df = pd.DataFrame(te_data, columns=te.columns_)

    print("\nTransaction Matrix:")
    display_html(df.head())

    # User Input for Minimum Support
    min_support = float(input("\nEnter Minimum Support Value (Example: 0.3): "))

    # Generate Frequent Itemsets
    frequent_itemsets = apriori(df,
                                min_support=min_support,
                                use_colnames=True)

    print("\n========== FREQUENT ITEMSETS ==========")

    print(frequent_itemsets)


    # User Input for Minimum Confidence
    min_confidence = float(input("\nEnter Minimum Confidence Value (Example: 0.7): "))

    # Generate Association Rules
    rules = association_rules(frequent_itemsets,
                              metric="confidence",
                              min_threshold=min_confidence)

    print("\n========== ASSOCIATION RULES ==========")

    if rules.empty:
        print("No Association Rules Found.")

    else:
        print(rules[['antecedents',
                     'consequents',
                     'support',
                     'confidence',
                     'lift']])


# Main Menu Function
def menu(dataset):

    print("""
    ========== APRIORI ALGORITHM MENU ==========

    1. View Transaction Dataset
    2. Apply Apriori Algorithm
    0. Exit
    """)

    option = int(input("Enter Your Option: "))

    while option != 0:

        if option == 1:
            view_transactions(dataset)

        elif option == 2:
            apply_apriori(dataset)

        else:
            print("Please Enter Correct Option.")

        option = int(input("\nEnter Your Option Again: "))

    print("\nProgram Ended Successfully.")


# Main Function
def main():

    # Sample Market Basket Dataset
    dataset = [

        ['Milk', 'Bread', 'Butter'],

        ['Bread', 'Butter'],

        ['Milk', 'Bread'],

        ['Milk', 'Butter'],

        ['Bread', 'Butter', 'Egg'],

        ['Milk', 'Egg'],

        ['Bread', 'Egg'],

        ['Milk', 'Bread', 'Butter', 'Egg']
    ]

    print("\n===================================================")
    print(" APRIORI ALGORITHM AND ASSOCIATION RULE GENERATION ")
    print("===================================================")

    print("\nPrepared By : Arjun Mijar")
    print("Roll No.    : 79010160")

    print("\nDo You Want to Start Program? (yes/no)")
    response = input()

    if response.lower() == 'yes':
        menu(dataset)

    else:
        print("\nProgram Terminated.")


# Calling Main Function
main()
```
