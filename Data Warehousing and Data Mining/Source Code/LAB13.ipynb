```python
# Demonstration of Clustering using DBSCAN Algorithm

# Prepared By : Arjun Mijar
# Roll No.    : 79010160

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN

from IPython.display import display_html

# Function to View Dataset
def view_dataset(data):

    print("\n========== DATASET ==========\n")

    display_html(data)

    print("\nNumber of Rows    :", data.shape[0])

    print("Number of Columns :", data.shape[1])


# Function to Apply DBSCAN Algorithm
def apply_dbscan(data):

    print("\n========== APPLYING DBSCAN ALGORITHM ==========\n")

    # Features for Clustering
    X = data[['Income', 'Spending']]

    # User Inputs
    eps_value = float(input("Enter EPS Value (Example: 10): "))

    min_points = int(input("Enter Minimum Points: "))

    # Create DBSCAN Model
    model = DBSCAN(eps=eps_value,
                   min_samples=min_points)

    # Fit Model
    data['Cluster'] = model.fit_predict(X)

    print("\n========== CLUSTERED DATA ==========\n")

    display_html(data)

    # Count Clusters
    total_clusters = len(set(model.labels_))

    print("\nTotal Clusters Found :", total_clusters)

    # Noise Points
    noise_points = list(model.labels_).count(-1)

    print("Noise Points :", noise_points)

    # Visualization
    print("\n========== CLUSTER VISUALIZATION ==========\n")

    plt.figure(figsize=(8,6))

    plt.scatter(data['Income'],
                data['Spending'],
                c=data['Cluster'],
                s=100)

    plt.xlabel("Income")

    plt.ylabel("Spending Score")

    plt.title("DBSCAN Clustering")

    plt.show()


# Menu Function
def menu(data):

    print("""
    ========== DBSCAN CLUSTERING MENU ==========

    1. View Dataset
    2. Apply DBSCAN Algorithm
    0. Exit
    """)

    option = int(input("Enter Your Option: "))

    while option != 0:

        if option == 1:
            view_dataset(data)

        elif option == 2:
            apply_dbscan(data)

        else:
            print("Please Enter Correct Option.")

        option = int(input("\nEnter Your Option Again: "))

    print("\nProgram Ended Successfully.")


# Main Function
def main():

    # Sample Dataset
    data = pd.DataFrame({

        'Customer': ['C1', 'C2', 'C3', 'C4',
                     'C5', 'C6', 'C7', 'C8'],

        'Income': [15, 16, 17, 40,
                   42, 43, 80, 85],

        'Spending': [39, 42, 40, 70,
                     72, 68, 20, 22]
    })

    print("\n===================================================")

    print(" CLUSTERING USING DBSCAN ALGORITHM ")

    print("===================================================")

    print("\nPrepared By : Arjun Mijar")

    print("Roll No.    : 79010160")

    print("\nDo You Want to Start Program? (yes/no)")

    response = input()

    if response.lower() == 'yes':

        menu(data)

    else:

        print("\nProgram Terminated.")


# Calling Main Function
main()
```
