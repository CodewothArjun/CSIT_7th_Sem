```python
# Demonstration of Classification using ID3 Algorithm

# Prepared By : Arjun Mijar
# Roll No.    : 79010160

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from IPython.display import display_html
import matplotlib.pyplot as plt

# Function to View Dataset
def view_dataset(data):

    print("\n========== DATASET ==========\n")

    display_html(data)

    print("\nNumber of Rows    :", data.shape[0])
    print("Number of Columns :", data.shape[1])


# Function to Apply ID3 Algorithm
def apply_id3(data):

    print("\n========== APPLYING ID3 ALGORITHM ==========\n")

    # Label Encoding
    encoder = LabelEncoder()

    data_encoded = data.copy()

    for column in data.columns:
        data_encoded[column] = encoder.fit_transform(data[column])

    print("Encoded Dataset:\n")
    display_html(data_encoded)

    # Features and Target
    X = data_encoded.drop('Play', axis=1)
    y = data_encoded['Play']

    # Create ID3 Model
    model = DecisionTreeClassifier(criterion='entropy')

    # Train Model
    model.fit(X, y)

    print("\nModel Trained Successfully.")

    # Predict New Data
    print("\n========== PREDICTION ==========\n")

    outlook = input("Enter Outlook (Sunny/Overcast/Rain): ")
    temperature = input("Enter Temperature (Hot/Mild/Cool): ")
    humidity = input("Enter Humidity (High/Normal): ")
    windy = input("Enter Windy (True/False): ")

    # Convert Input into Encoded Form
    input_data = pd.DataFrame({
        'Outlook': [outlook],
        'Temperature': [temperature],
        'Humidity': [humidity],
        'Windy': [windy]
    })

    # Encode Input Data
    for column in input_data.columns:
        input_data[column] = LabelEncoder().fit(data[column]).transform(input_data[column])

    # Prediction
    prediction = model.predict(input_data)

    print("\nPrediction Result:")

    if prediction[0] == 1:
        print("Play Tennis : Yes")
    else:
        print("Play Tennis : No")

    # Display Decision Tree
    print("\n========== DECISION TREE ==========\n")

    plt.figure(figsize=(12,8))

    tree.plot_tree(model,
                   feature_names=X.columns,
                   class_names=['No', 'Yes'],
                   filled=True)

    plt.show()


# Menu Function
def menu(data):

    print("""
    ========== ID3 ALGORITHM MENU ==========

    1. View Dataset
    2. Apply ID3 Algorithm
    0. Exit
    """)

    option = int(input("Enter Your Option: "))

    while option != 0:

        if option == 1:
            view_dataset(data)

        elif option == 2:
            apply_id3(data)

        else:
            print("Please Enter Correct Option.")

        option = int(input("\nEnter Your Option Again: "))

    print("\nProgram Ended Successfully.")


# Main Function
def main():

    # Play Tennis Dataset
    data = pd.DataFrame({

        'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain',
                    'Rain', 'Rain', 'Overcast', 'Sunny',
                    'Sunny', 'Rain', 'Sunny', 'Overcast',
                    'Overcast', 'Rain'],

        'Temperature': ['Hot', 'Hot', 'Hot', 'Mild',
                        'Cool', 'Cool', 'Cool', 'Mild',
                        'Cool', 'Mild', 'Mild', 'Mild',
                        'Hot', 'Mild'],

        'Humidity': ['High', 'High', 'High', 'High',
                     'Normal', 'Normal', 'Normal', 'High',
                     'Normal', 'Normal', 'Normal', 'High',
                     'Normal', 'High'],

        'Windy': ['False', 'True', 'False', 'False',
                  'False', 'True', 'True', 'False',
                  'False', 'False', 'True', 'True',
                  'False', 'True'],

        'Play': ['No', 'No', 'Yes', 'Yes',
                 'Yes', 'No', 'Yes', 'No',
                 'Yes', 'Yes', 'Yes', 'Yes',
                 'Yes', 'No']
    })

    print("\n==============================================")
    print(" CLASSIFICATION USING ID3 ALGORITHM ")
    print("==============================================")

    print("\nPrepared By : Arjun Mijar")
    print("Roll No.    : 79010160")

    print("\nDo You Want to Start Program? (yes/no)")
    response = input()

    if response.lower() == 'yes':
        menu(data)

    else:
        print("\nProgram Terminated.")


# Calling Main Function
main()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from IPython.display import display
import matplotlib.pyplot as plt

class ID3Classifier:
    def __init__(self):
        self.encoders = {}
        self.model = None
        self.feature_names = None
        
    def view_dataset(self, data):
        print("\n========== DATASET ==========\n")
        display(data)
        print(f"\nNumber of Rows    : {data.shape[0]}")
        print(f"Number of Columns : {data.shape[1]}")

    def apply_id3(self, data):
        print("\n========== APPLYING ID3 ALGORITHM ==========\n")
        
        # Store encoders for later use
        data_encoded = data.copy()
        
        for column in data.columns:
            encoder = LabelEncoder()
            data_encoded[column] = encoder.fit_transform(data[column])
            self.encoders[column] = encoder
        
        print("Encoded Dataset:\n")
        display(data_encoded)
        
        # Features and Target
        X = data_encoded.drop('Play', axis=1)
        y = data_encoded['Play']
        self.feature_names = X.columns
        
        # Create and train ID3 Model
        self.model = DecisionTreeClassifier(criterion='entropy', random_state=42)
        self.model.fit(X, y)
        
        print("\nModel Trained Successfully.")
        
        # Make prediction
        self.predict_new_data(data)
        
        # Display decision tree
        self.display_tree()

    def predict_new_data(self, original_data):
        print("\n========== PREDICTION ==========\n")
        
        try:
            outlook = input("Enter Outlook (Sunny/Overcast/Rain): ")
            temperature = input("Enter Temperature (Hot/Mild/Cool): ")
            humidity = input("Enter Humidity (High/Normal): ")
            windy = input("Enter Windy (True/False): ")
            
            # Create input dataframe
            input_data = pd.DataFrame({
                'Outlook': [outlook],
                'Temperature': [temperature],
                'Humidity': [humidity],
                'Windy': [windy]
            })
            
            # Encode using stored encoders
            input_encoded = input_data.copy()
            for column in input_data.columns:
                input_encoded[column] = self.encoders[column].transform(input_data[column])
            
            # Make prediction
            prediction = self.model.predict(input_encoded)
            
            print("\nPrediction Result:")
            result = "Yes" if prediction[0] == 1 else "No"
            print(f"Play Tennis : {result}")
            
        except ValueError as e:
            print(f"Error: Invalid input value. Please enter valid options. {e}")
        except Exception as e:
            print(f"An error occurred: {e}")

    def display_tree(self):
        print("\n========== DECISION TREE ==========\n")
        
        plt.figure(figsize=(15, 10))
        tree.plot_tree(self.model,
                      feature_names=self.feature_names,
                      class_names=['No', 'Yes'],
                      filled=True,
                      rounded=True,
                      fontsize=10)
        plt.title("ID3 Decision Tree for Play Tennis Dataset")
        plt.show()

def menu():
    # Play Tennis Dataset
    data = pd.DataFrame({
        'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 
                   'Overcast', 'Sunny', 'Sunny', 'Rain', 'Sunny', 'Overcast', 
                   'Overcast', 'Rain'],
        'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 
                       'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
        'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 
                    'Normal', 'High', 'Normal', 'Normal', 'Normal', 'High', 
                    'Normal', 'High'],
        'Windy': ['False', 'True', 'False', 'False', 'False', 'True', 'True', 
                 'False', 'False', 'False', 'True', 'True', 'False', 'True'],
        'Play': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 
                'Yes', 'Yes', 'Yes', 'Yes', 'No']
    })
    
    classifier = ID3Classifier()
    
    print("\n==============================================")
    print(" CLASSIFICATION USING ID3 ALGORITHM ")
    print("==============================================")
    
    while True:
        print("""
    ========== ID3 ALGORITHM MENU ==========

    1. View Dataset
    2. Apply ID3 Algorithm
    0. Exit
        """)
        
        try:
            option = int(input("Enter Your Option: "))
            
            if option == 0:
                print("\nProgram Ended Successfully.")
                break
            elif option == 1:
                classifier.view_dataset(data)
            elif option == 2:
                classifier.apply_id3(data)
            else:
                print("Please Enter Correct Option (0, 1, or 2).")
                
        except ValueError:
            print("Please enter a valid number.")
        except KeyboardInterrupt:
            print("\n\nProgram interrupted by user.")
            break

if __name__ == "__main__":
    menu()
```
