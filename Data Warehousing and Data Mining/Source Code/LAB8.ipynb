```python
# Implementation of Linear Regression Algorithm

# Prepared By : Arjun Mijar
# Roll No.    : 79010160

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from IPython.display import display_html

# Function to View Dataset
def view_dataset(data):
    print("\n========== DATASET ==========\n")
    display_html(data)
    print("\nNumber of Rows    :", data.shape[0])
    print("Number of Columns :", data.shape[1])

# Function to Apply Linear Regression
def apply_linear_regression(data):
    print("\n========== APPLYING LINEAR REGRESSION ==========\n")

    # Feature and Target Variable
    X = data[['Hours']]
    y = data['Marks']

    # Create Model
    model = LinearRegression()

    # Train Model
    model.fit(X, y)
    print("Model Trained Successfully.")

    # Regression Equation
    print("\nRegression Equation:")
    print(f"Marks = {model.coef_[0]:.2f} * Hours + {model.intercept_:.2f}")

    # Predict Existing Data
    y_pred = model.predict(X)

    # Performance Evaluation
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)

    print("\nModel Evaluation:")
    print("Mean Squared Error :", round(mse, 2))
    print("R2 Score           :", round(r2, 2))

    # User Prediction - FIX: Use DataFrame format for consistency
    print("\n========== PREDICTION ==========\n")
    hours = float(input("Enter Study Hours: "))
    
    # Create DataFrame with same column name as training data
    hours_df = pd.DataFrame({'Hours': [hours]})
    predicted_marks = model.predict(hours_df)
    
    print("\nPredicted Marks :", round(predicted_marks[0], 2))

    # Visualization
    print("\n========== REGRESSION GRAPH ==========\n")
    plt.figure(figsize=(8,5))
    plt.scatter(X, y, label='Actual Data')
    plt.plot(X, y_pred, linewidth=2)
    plt.xlabel("Study Hours")
    plt.ylabel("Marks")
    plt.title("Linear Regression")
    plt.show()

# Menu Function
def menu(data):
    print("""
    ========== LINEAR REGRESSION MENU ==========

    1. View Dataset
    2. Apply Linear Regression
    0. Exit
    """)

    option = int(input("Enter Your Option: "))

    while option != 0:
        if option == 1:
            view_dataset(data)
        elif option == 2:
            apply_linear_regression(data)
        else:
            print("Please Enter Correct Option.")
        option = int(input("\nEnter Your Option Again: "))

    print("\nProgram Ended Successfully.")

# Main Function
def main():
    # Sample Dataset
    data = pd.DataFrame({
        'Hours': [1, 2, 3, 4, 5, 6, 7, 8],
        'Marks': [35, 40, 50, 55, 60, 65, 70, 80]
    })

    print("\n===================================================")
    print(" IMPLEMENTATION OF LINEAR REGRESSION ALGORITHM ")
    print("===================================================")
    print("\nPrepared By : Arjun Mijar")
    print("Roll No.    : 79010160")
    print("\nDo You Want to Start Program? (yes/no)")
    response = input()

    if response.lower() == 'yes':
        menu(data)
    else:
        print("\nProgram Terminated.")

# Calling Main Function
main()
```
