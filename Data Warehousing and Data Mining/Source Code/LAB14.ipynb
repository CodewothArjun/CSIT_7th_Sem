```python
# Demonstration of Classification using Backpropagation Algorithm

# Prepared By : Arjun Mijar
# Roll No.    : 79010160

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from IPython.display import display_html

# Function to View Dataset
def view_dataset(data):

    print("\n========== DATASET ==========\n")

    display_html(data)

    print("\nNumber of Rows    :", data.shape[0])

    print("Number of Columns :", data.shape[1])


# Function to Apply Backpropagation Algorithm
def apply_backpropagation(data):

    print("\n========== APPLYING BACKPROPAGATION ALGORITHM ==========\n")

    # Encode Dataset
    encoder = LabelEncoder()

    data_encoded = data.copy()

    for column in data.columns:
        data_encoded[column] = encoder.fit_transform(data[column])

    print("Encoded Dataset:\n")

    display_html(data_encoded)

    # Features and Target Variable
    X = data_encoded.drop('Play', axis=1)

    y = data_encoded['Play']

    # Create Neural Network Model
    model = MLPClassifier(hidden_layer_sizes=(5,),
                          max_iter=1000,
                          random_state=1)

    # Train Model
    model.fit(X, y)

    print("\nModel Trained Successfully.")

    # Training Accuracy
    prediction_train = model.predict(X)

    accuracy = accuracy_score(y, prediction_train)

    print("\nTraining Accuracy :", round(accuracy * 100, 2), "%")

    # User Prediction
    print("\n========== PREDICTION ==========\n")

    outlook = input("Enter Outlook (Sunny/Overcast/Rain): ")

    temperature = input("Enter Temperature (Hot/Mild/Cool): ")

    humidity = input("Enter Humidity (High/Normal): ")

    windy = input("Enter Windy (True/False): ")

    # Create Input Data
    input_data = pd.DataFrame({

        'Outlook': [outlook],

        'Temperature': [temperature],

        'Humidity': [humidity],

        'Windy': [windy]
    })

    # Encode Input Data
    for column in input_data.columns:
        input_data[column] = LabelEncoder().fit(data[column]).transform(input_data[column])

    # Prediction
    prediction = model.predict(input_data)

    print("\n========== PREDICTION RESULT ==========\n")

    if prediction[0] == 1:

        print("Play Tennis : Yes")

    else:

        print("Play Tennis : No")


# Menu Function
def menu(data):

    print("""
    ========== BACKPROPAGATION MENU ==========

    1. View Dataset
    2. Apply Backpropagation Algorithm
    0. Exit
    """)

    option = int(input("Enter Your Option: "))

    while option != 0:

        if option == 1:
            view_dataset(data)

        elif option == 2:
            apply_backpropagation(data)

        else:
            print("Please Enter Correct Option.")

        option = int(input("\nEnter Your Option Again: "))

    print("\nProgram Ended Successfully.")


# Main Function
def main():

    # Play Tennis Dataset
    data = pd.DataFrame({

        'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain',
                    'Rain', 'Rain', 'Overcast', 'Sunny',
                    'Sunny', 'Rain', 'Sunny', 'Overcast',
                    'Overcast', 'Rain'],

        'Temperature': ['Hot', 'Hot', 'Hot', 'Mild',
                        'Cool', 'Cool', 'Cool', 'Mild',
                        'Cool', 'Mild', 'Mild', 'Mild',
                        'Hot', 'Mild'],

        'Humidity': ['High', 'High', 'High', 'High',
                     'Normal', 'Normal', 'Normal', 'High',
                     'Normal', 'Normal', 'Normal', 'High',
                     'Normal', 'High'],

        'Windy': ['False', 'True', 'False', 'False',
                  'False', 'True', 'True', 'False',
                  'False', 'False', 'True', 'True',
                  'False', 'True'],

        'Play': ['No', 'No', 'Yes', 'Yes',
                 'Yes', 'No', 'Yes', 'No',
                 'Yes', 'Yes', 'Yes', 'Yes',
                 'Yes', 'No']
    })

    print("\n========================================================")

    print(" CLASSIFICATION USING BACKPROPAGATION ALGORITHM ")

    print("========================================================")

    print("\nPrepared By : Arjun Mijar")

    print("Roll No.    : 79010160")

    print("\nDo You Want to Start Program? (yes/no)")

    response = input()

    if response.lower() == 'yes':

        menu(data)

    else:

        print("\nProgram Terminated.")


# Calling Main Function
main()
```
