```python
# Demonstration of Clustering using Agglomerative Algorithm

# Prepared By : Arjun Mijar
# Roll No.    : 79010160

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import AgglomerativeClustering

from scipy.cluster.hierarchy import dendrogram, linkage

from IPython.display import display_html

# Function to View Dataset
def view_dataset(data):

    print("\n========== DATASET ==========\n")

    display_html(data)

    print("\nNumber of Rows    :", data.shape[0])

    print("Number of Columns :", data.shape[1])


# Function to Apply Agglomerative Clustering
def apply_agglomerative(data):

    print("\n========== APPLYING AGGLOMERATIVE CLUSTERING ==========\n")

    # Features for Clustering
    X = data[['Income', 'Spending']]

    # User Input for Number of Clusters
    k = int(input("Enter Number of Clusters: "))

    # Create Agglomerative Model
    model = AgglomerativeClustering(n_clusters=k)

    # Fit Model
    data['Cluster'] = model.fit_predict(X)

    print("\n========== CLUSTERED DATA ==========\n")

    display_html(data)

    # Visualization of Clusters
    print("\n========== CLUSTER VISUALIZATION ==========\n")

    plt.figure(figsize=(8,6))

    plt.scatter(data['Income'],
                data['Spending'],
                c=data['Cluster'],
                s=100)

    plt.xlabel("Income")

    plt.ylabel("Spending Score")

    plt.title("Agglomerative Clustering")

    plt.show()

    # Dendrogram
    print("\n========== DENDROGRAM ==========\n")

    linked = linkage(X, method='ward')

    plt.figure(figsize=(10,6))

    dendrogram(linked,
               labels=data['Customer'].tolist())

    plt.title("Dendrogram")

    plt.xlabel("Customers")

    plt.ylabel("Distance")

    plt.show()


# Menu Function
def menu(data):

    print("""
    ========== AGGLOMERATIVE CLUSTERING MENU ==========

    1. View Dataset
    2. Apply Agglomerative Clustering
    0. Exit
    """)

    option = int(input("Enter Your Option: "))

    while option != 0:

        if option == 1:
            view_dataset(data)

        elif option == 2:
            apply_agglomerative(data)

        else:
            print("Please Enter Correct Option.")

        option = int(input("\nEnter Your Option Again: "))

    print("\nProgram Ended Successfully.")


# Main Function
def main():

    # Sample Dataset
    data = pd.DataFrame({

        'Customer': ['C1', 'C2', 'C3', 'C4',
                     'C5', 'C6', 'C7', 'C8'],

        'Income': [15, 16, 17, 40,
                   42, 43, 80, 85],

        'Spending': [39, 42, 40, 70,
                     72, 68, 20, 22]
    })

    print("\n========================================================")

    print(" CLUSTERING USING AGGLOMERATIVE ALGORITHM ")

    print("========================================================")

    print("\nPrepared By : Arjun Mijar")

    print("Roll No.    : 79010160")

    print("\nDo You Want to Start Program? (yes/no)")

    response = input()

    if response.lower() == 'yes':

        menu(data)

    else:

        print("\nProgram Terminated.")


# Calling Main Function
main()
```
