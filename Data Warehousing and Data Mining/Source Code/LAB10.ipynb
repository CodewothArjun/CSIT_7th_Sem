```python
# Demonstration of Clustering using K-Means Algorithm

# Prepared By : Arjun Mijar
# Roll No.    : 79010160

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from IPython.display import display_html

# Function to View Dataset
def view_dataset(data):

    print("\n========== DATASET ==========\n")

    display_html(data)

    print("\nNumber of Rows    :", data.shape[0])

    print("Number of Columns :", data.shape[1])


# Function to Apply K-Means Algorithm
def apply_kmeans(data):

    print("\n========== APPLYING K-MEANS ALGORITHM ==========\n")

    # Features for Clustering
    X = data[['Income', 'Spending']]

    # User Input for Number of Clusters
    k = int(input("Enter Number of Clusters (K): "))

    # Create K-Means Model
    model = KMeans(n_clusters=k, random_state=0)

    # Train Model
    model.fit(X)

    # Predict Cluster Labels
    data['Cluster'] = model.labels_

    print("\n========== CLUSTERED DATA ==========\n")

    display_html(data)

    # Display Centroids
    print("\nCluster Centroids:\n")

    centroids = pd.DataFrame(model.cluster_centers_,
                             columns=['Income', 'Spending'])

    print(centroids)

    # Visualization
    print("\n========== CLUSTER VISUALIZATION ==========\n")

    plt.figure(figsize=(8,6))

    plt.scatter(data['Income'],
                data['Spending'],
                c=data['Cluster'],
                s=100)

    # Plot Centroids
    plt.scatter(model.cluster_centers_[:, 0],
                model.cluster_centers_[:, 1],
                marker='X',
                s=300)

    plt.xlabel("Income")

    plt.ylabel("Spending Score")

    plt.title("K-Means Clustering")

    plt.show()


# Menu Function
def menu(data):

    print("""
    ========== K-MEANS MENU ==========

    1. View Dataset
    2. Apply K-Means Algorithm
    0. Exit
    """)

    option = int(input("Enter Your Option: "))

    while option != 0:

        if option == 1:
            view_dataset(data)

        elif option == 2:
            apply_kmeans(data)

        else:
            print("Please Enter Correct Option.")

        option = int(input("\nEnter Your Option Again: "))

    print("\nProgram Ended Successfully.")


# Main Function
def main():

    # Sample Dataset
    data = pd.DataFrame({

        'Customer': ['C1', 'C2', 'C3', 'C4',
                     'C5', 'C6', 'C7', 'C8'],

        'Income': [15, 16, 17, 40,
                   42, 43, 80, 85],

        'Spending': [39, 42, 40, 70,
                     72, 68, 20, 22]
    })

    print("\n===================================================")

    print(" CLUSTERING USING K-MEANS ALGORITHM ")

    print("===================================================")

    print("\nPrepared By : Arjun Mijar")

    print("Roll No.    : 79010160")

    print("\nDo You Want to Start Program? (yes/no)")

    response = input()

    if response.lower() == 'yes':

        menu(data)

    else:

        print("\nProgram Terminated.")


# Calling Main Function
main()
```

```python
# Demonstration of Clustering using K-Means Algorithm

# Prepared By : Arjun Mijar
# Roll No.    : 79010160

# Fix for KMeans memory leak warning on Windows - MUST be at the very top
import os
import warnings

# Set environment variables before any sklearn imports
os.environ['OMP_NUM_THREADS'] = '1'
os.environ['MKL_NUM_THREADS'] = '1'

# Suppress the specific sklearn warning
warnings.filterwarnings('ignore', message='KMeans is known to have a memory leak on Windows with MKL')

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from IPython.display import display_html

# Function to View Dataset
def view_dataset(data):

    print("\n========== DATASET ==========\n")

    display_html(data)

    print("\nNumber of Rows    :", data.shape[0])

    print("Number of Columns :", data.shape[1])


# Function to Apply K-Means Algorithm
def apply_kmeans(data):

    print("\n========== APPLYING K-MEANS ALGORITHM ==========\n")

    # Features for Clustering
    X = data[['Income', 'Spending']]

    # User Input for Number of Clusters
    k = int(input("Enter Number of Clusters (K): "))

    # Create K-Means Model with additional parameters to prevent warnings
    model = KMeans(n_clusters=k, random_state=0, n_init=10, algorithm='lloyd')

    # Train Model
    model.fit(X)

    # Predict Cluster Labels
    data['Cluster'] = model.labels_

    print("\n========== CLUSTERED DATA ==========\n")

    display_html(data)

    # Display Centroids
    print("\nCluster Centroids:\n")

    centroids = pd.DataFrame(model.cluster_centers_,
                             columns=['Income', 'Spending'])

    print(centroids)

    # Visualization
    print("\n========== CLUSTER VISUALIZATION ==========\n")

    plt.figure(figsize=(8,6))

    # Create scatter plot with different colors for each cluster
    scatter = plt.scatter(data['Income'],
                         data['Spending'],
                         c=data['Cluster'],
                         s=100,
                         cmap='viridis',
                         alpha=0.7)

    # Plot Centroids
    plt.scatter(model.cluster_centers_[:, 0],
                model.cluster_centers_[:, 1],
                marker='X',
                s=300,
                c='red',
                edgecolors='black',
                linewidth=2,
                label='Centroids')

    plt.xlabel("Income")

    plt.ylabel("Spending Score")

    plt.title("K-Means Clustering")
    
    plt.colorbar(scatter, label='Cluster')
    
    plt.legend()

    plt.grid(True, alpha=0.3)

    plt.show()


# Menu Function
def menu(data):

    print("""
    ========== K-MEANS MENU ==========

    1. View Dataset
    2. Apply K-Means Algorithm
    0. Exit
    """)

    option = int(input("Enter Your Option: "))

    while option != 0:

        if option == 1:
            view_dataset(data)

        elif option == 2:
            # Create a copy of data to avoid modifying original
            data_copy = data.copy()
            apply_kmeans(data_copy)

        else:
            print("Please Enter Correct Option.")

        option = int(input("\nEnter Your Option Again: "))

    print("\nProgram Ended Successfully.")


# Main Function
def main():

    # Sample Dataset
    data = pd.DataFrame({

        'Customer': ['C1', 'C2', 'C3', 'C4',
                     'C5', 'C6', 'C7', 'C8'],

        'Income': [15, 16, 17, 40,
                   42, 43, 80, 85],

        'Spending': [39, 42, 40, 70,
                     72, 68, 20, 22]
    })

    print("\n===================================================")

    print(" CLUSTERING USING K-MEANS ALGORITHM ")

    print("===================================================")

    print("\nPrepared By : Arjun Mijar")

    print("Roll No.    : 79010160")

    print("\nDo You Want to Start Program? (yes/no)")

    response = input()

    if response.lower() == 'yes':

        menu(data)

    else:

        print("\nProgram Terminated.")


# Calling Main Function
if __name__ == "__main__":
    main()
```
