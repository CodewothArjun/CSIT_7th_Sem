```python
# Demonstration of Data Preprocessing and Cleaning using Python Libraries

# Prepared By: Arjun Mijar
# Roll No.: 79010160

import pandas as pd
import numpy as np
from IPython.display import display_html

# Function to View Dataset
def view_data(data):
    print("\n========== DATASET INFORMATION ==========")
    print("Number of Rows    :", data.shape[0])
    print("Number of Columns :", data.shape[1])

    print("\n========== FIRST FIVE RECORDS ==========")
    display_html(data.head())


# Function to Replace Missing Values
def replace_missing_values(data):

    print("\n========== REPLACING MISSING VALUES ==========")

    data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
    data['Salary'] = pd.to_numeric(data['Salary'], errors='coerce')

    print("\nMissing Values Before Cleaning:")
    print(data.isnull().sum())

    # Replace by Mean
    data['Age'] = data['Age'].fillna(data['Age'].mean())
    data['Salary'] = data['Salary'].fillna(data['Salary'].mean())

    print("\nMissing Values After Cleaning:")
    print(data.isnull().sum())

    print("\nUpdated Dataset:")
    display_html(data.head())


# Function to Remove Duplicate Records
def remove_duplicates(data):

    print("\n========== DUPLICATE RECORD HANDLING ==========")

    duplicates = data.duplicated()

    print("Number of Duplicate Rows :", duplicates.sum())

    print("Rows Before Removing Duplicates :", data.shape[0])

    data2 = data.drop_duplicates()

    print("Rows After Removing Duplicates  :", data2.shape[0])


# Function for Sampling
def sampling(data):

    print("\n========== SAMPLING ==========")

    print("\nOriginal Dataset:")
    display_html(data.head())

    print("\nRandom Sample of 3 Rows:")
    sample = data.sample(n=3)
    display_html(sample)

    print("\nRandom 30% Sample:")
    sample = data.sample(frac=0.3, random_state=1)
    display_html(sample)


# Function for Discretization
def discretization(data):

    print("\n========== DISCRETIZATION ==========")

    print("\nDistribution of Age Values:")
    print(data['Age'].value_counts(sort=False))

    print("\nEqual Width Binning:")
    bins = pd.cut(data['Age'], 4)

    print(bins.value_counts(sort=False))

    print("\nEqual Frequency Binning:")
    bins = pd.qcut(data['Age'], 4)

    print(bins.value_counts(sort=False))


# Function to Handle Outliers
def outlier_detection(data):

    print("\n========== OUTLIER DETECTION ==========")

    data2 = data[['Age', 'Salary']]

    Z = (data2 - data2.mean()) / data2.std()

    print("\nZ-Score Values:")
    print(Z.head())

    print("\nRows Before Removing Outliers :", Z.shape[0])

    Z2 = Z.loc[
        ((Z > -3).sum(axis=1) == 2) &
        ((Z <= 3).sum(axis=1) == 2), :
    ]

    print("Rows After Removing Outliers  :", Z2.shape[0])


# Main Preprocessing Menu
def preprocessing(data):

    print("\n========== DATA PREPROCESSING MENU ==========")

    print("""
    0. Exit
    1. View Dataset
    2. Replace Missing Values
    3. Remove Duplicate Rows
    4. Sampling
    5. Discretization
    6. Outlier Detection
    """)

    option = int(input("Enter Your Option: "))

    while option != 0:

        if option == 1:
            view_data(data)

        elif option == 2:
            replace_missing_values(data)

        elif option == 3:
            remove_duplicates(data)

        elif option == 4:
            sampling(data)

        elif option == 5:
            discretization(data)

        elif option == 6:
            outlier_detection(data)

        else:
            print("Please Enter Correct Option.")

        option = int(input("\nEnter Your Option Again: "))

    print("\nProgram Ended Successfully.")


# Main Function
def main():

    # Sample Dataset
    data = pd.DataFrame({

        'Name': ['Ram', 'Shyam', 'Hari', 'Sita', 'Gita', 'Ram'],

        'Age': [23, 25, np.nan, 22, 24, 23],

        'Salary': [50000, 60000, 55000, np.nan, 52000, 50000],

        'Gender': ['Male', 'Male', 'Male', 'Female', 'Female', 'Male']
    })

    print("\n==============================================")
    print(" DATA PREPROCESSING AND CLEANING USING PYTHON ")
    print("==============================================")

    print("\nPrepared By : Arjun Mijar")
    print("Roll No.    : 79010160")

    print("\nDo You Want to Start Program? (yes/no)")
    response = input()

    if response.lower() == 'yes':
        preprocessing(data)

    else:
        print("\nProgram Terminated.")


# Calling Main Function
main()
```
