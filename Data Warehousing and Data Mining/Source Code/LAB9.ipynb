```python
# Implementation of Multiple Linear Regression Algorithm

# Prepared By : Arjun Mijar
# Roll No.    : 79010160

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from IPython.display import display_html

# Function to View Dataset
def view_dataset(data):
    print("\n========== DATASET ==========\n")
    display_html(data)
    print("\nNumber of Rows    :", data.shape[0])
    print("Number of Columns :", data.shape[1])

# Function to Apply Multiple Linear Regression
def apply_multiple_regression(data):
    print("\n========== APPLYING MULTIPLE LINEAR REGRESSION ==========\n")
    
    # Features and Target Variable
    X = data[['Hours', 'Attendance']]
    y = data['Marks']
    
    # Create Model
    model = LinearRegression()
    
    # Train Model
    model.fit(X, y)
    print("Model Trained Successfully.")
    
    # Display Coefficients
    print("\nRegression Equation:")
    print(f"""
Marks = ({model.coef_[0]:.2f} * Hours)
      + ({model.coef_[1]:.2f} * Attendance)
      + ({model.intercept_:.2f})
""")
    
    # Predict Existing Data
    y_pred = model.predict(X)
    
    # Model Evaluation
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    
    print("========== MODEL EVALUATION ==========\n")
    print("Mean Squared Error :", round(mse, 2))
    print("R2 Score           :", round(r2, 2))
    
    # User Prediction
    print("\n========== PREDICTION ==========\n")
    hours = float(input("Enter Study Hours: "))
    attendance = float(input("Enter Attendance Percentage: "))
    
    # FIX: Create DataFrame with proper feature names for prediction
    user_input = pd.DataFrame({
        'Hours': [hours],
        'Attendance': [attendance]
    })
    
    prediction = model.predict(user_input)
    print("\nPredicted Marks :", round(prediction[0], 2))

# Menu Function
def menu(data):
    print("""
    ========== MULTIPLE LINEAR REGRESSION MENU ==========

    1. View Dataset
    2. Apply Multiple Linear Regression
    0. Exit
    """)
    
    option = int(input("Enter Your Option: "))
    
    while option != 0:
        if option == 1:
            view_dataset(data)
        elif option == 2:
            apply_multiple_regression(data)
        else:
            print("Please Enter Correct Option.")
        
        option = int(input("\nEnter Your Option Again: "))
    
    print("\nProgram Ended Successfully.")

# Main Function
def main():
    # Sample Dataset
    data = pd.DataFrame({
        'Hours': [1, 2, 3, 4, 5, 6, 7, 8],
        'Attendance': [60, 65, 70, 72, 78, 82, 88, 95],
        'Marks': [30, 35, 45, 50, 60, 65, 75, 85]
    })
    
    print("\n========================================================")
    print(" IMPLEMENTATION OF MULTIPLE LINEAR REGRESSION ")
    print("========================================================")
    print("\nPrepared By : Arjun Mijar")
    print("Roll No.    : 79010160")
    print("\nDo You Want to Start Program? (yes/no)")
    
    response = input()
    
    if response.lower() == 'yes':
        menu(data)
    else:
        print("\nProgram Terminated.")

# Calling Main Function
main()
```
